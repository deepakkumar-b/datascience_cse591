```python
import matplotlib.pyplot as plt
import prettyplotlib as ppl
import matplotlib as mpl
from pandas import DataFrame

#Convert the latitude and logitude to projectable system
def mapping(old):
    direction = {'N':-1, 'S':1, 'E': -1, 'W':1}
    new = old.split()
    new_dir = new.pop()
    new.extend([0,0,0])
    return (int(new[0])+int(new[1])/60.0+int(new[2])/3600.0) * direction[new_dir]

def get_marker_color(population):
    # Returns red for very high usage, blue for high usage, 
    # yellow for moderate and green for low usage.
    if population > 0 and population < 99999:
        return ('go')
    elif population > 99999 and population < 99999:
        return ('yo')
    elif population > 99999 and population < 999999:
        return ('bo')
    else:
        return ('ro')
    
# make sure the value of resolution is a lowercase L,
#  for 'low', not a numeral 1
map = Basemap(projection='robin', lat_0=0, lon_0=0,
              resolution='c', area_thresh=1000.0)
 

land_color = 'lightgreen'
water_color = 'blue'

map.fillcontinents(color=land_color, lake_color=water_color)    
map.drawcoastlines()
map.drawcountries()
#map.bluemarble()
#map.fillcontinents(color='coral')
map.drawmapboundary()
 
map.drawmeridians(np.arange(0, 360, 30))
map.drawparallels(np.arange(-90, 90, 30))

#Read latitides, Longitudes and Internet Users
lats = dfCountries['Latitude of Capital']
lons = dfCountries['Longitude of Capital']
population = dfCountries['Labor Force']


#Adding markers to the map
min_marker_size = 5
for lon, lat, pop in zip(lons, lats, population):
    x,y = map(mapping(str(lon)), mapping(str(lat)))
    #x,y = map(mapping(int(lon)), mapping(int(lat)))
    marker_color = get_marker_color(pop)
    #msize = iuser * (min_marker_size/1000)
    map.plot(x, y, marker_color, markersize=min_marker_size)
    #map.plot()
#lon = lons.flat[1]#-135.3318
#lat = lats.flat[1]#57.0799
#x,y = map(lon, lat)
#map.plot(x, y, 'bo', markersize=24)
figsize(150, 150)
plt.show()

```

```python
#Histogram of Life Expectancy and Literacy Rate
fig, axes = plt.subplots(1, 2, figsize=(12,4))
#Health Expenditure/GDP
health = dfCountries['Health Expenditure/GDP']
N=len(health)
x = np.arange(N)
area =  np.pi * (8 * health/10)**2
colors = np.random.rand(N)
ppl.scatter(axes[0],x,health,s=area, c=colors, alpha=0.5)
axes[0].set_title("Health Expenditure/GDP - Scatter Plot\n (Area of the circle proportional\n to the Health Expenditure/GDP)\n")
axes[0].set_ylabel('Health Expenditure/GDP')
axes[0].set_xlabel('Countries')

#Net Migration Rate/1000 Population
netMigration = dfCountries['Net Migration Rate/1000 Population']
N = len(netMigration)
x = np.arange(N)
area =  np.pi * (8 * netMigration+50)**2
colors = np.random.rand(N)
ppl.scatter(axes[1],x,netMigration,s=area/600, c=colors, alpha=0.5)
axes[1].set_title("Net Migration Rate/1000 Population - Scatter Plot\n (Area of the circle proportional\n to the Net Migration Rate/1000 Population)\n")
axes[1].set_ylabel('Net Migration Rate/1000 Population')
axes[1].set_xlabel('Countries')
```
