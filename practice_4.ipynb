```python
import pandas as pd
from scipy import stats
from pandas import Series, DataFrame
```

```python
#Read csv File into dataFrame
Countries = pd.read_csv('countries.csv')

#Arbitrary assignment of weights
pop_area = -5
int_pop = 10
pop_area_lit = 45
pop_lit = 40
lab_pop = 35
population = -5
life = 8
gdp = 20
coast = 8
literacy = 15
water = 15
labor = 15
migration = -5


def sendforScoring(col,wt):
    #Normalize the data frame
    Countries['z'+col] = stats.zscore(Countries[col])
    #shifting rightwards to make it non negative
    Countries['z'+col] = Countries['z'+col] + (-1 * min(Countries['z'+col]))
    #Projecting the values between 0-1
    Countries['z'+col] = Countries['z'+col] / max(Countries['z'+col])
    #Weighted Score
    Countries['z'+col] = literacy * Countries['z'+col]
    #print(Countries['z'+colName])


#Add the needed extra columns
Countries['Population/Area*Literacy'] = Countries['Literacy Rate']*Countries['Population']/Countries['Area (sq km)']
sendforScoring('Population/Area*Literacy',pop_area_lit)

Countries['Population*Literacy'] = Countries['Literacy Rate']*Countries['Population']
sendforScoring('Population*Literacy',pop_lit)
#fair distribution
Countries['Population/Area'] = Countries['Population']/Countries['Area (sq km)']
sendforScoring('Population/Area',pop_area)

Countries['Labor Force/Population'] = Countries['Labor Force']/Countries['Population']
sendforScoring('Labor Force/Population',lab_pop)

Countries['Internet Users/Population'] = Countries['Internet Users (circa 2009)']/Countries['Population']
sendforScoring('Internet Users/Population',int_pop)

#Normalize the data
sendforScoring('Population',population)
sendforScoring('Life Expectancy',life)
sendforScoring('GDP (PPP) in US $',gdp)
sendforScoring('Coastline (km)',coast)
sendforScoring('Literacy Rate',literacy)
sendforScoring('Renewable Water Resources (cu km)',water)
sendforScoring('Labor Force',labor)
sendforScoring('Net Migration Rate/1000 Population',migration)


#Compute the netScore
Countries['Score'] = Countries['zPopulation/Area'] + Countries['zLife Expectancy'] + Countries['zPopulation/Area*Literacy']
Countries['Score'] = Countries['Score'] + Countries['zCoastline (km)'] 
Countries['Score'] = Countries['Score'] + Countries['zGDP (PPP) in US $']
Countries['Score'] = Countries['Score'] + Countries['zLiteracy Rate']
Countries['Score'] = Countries['Score'] + Countries['zRenewable Water Resources (cu km)'] + Countries['zLabor Force']
Countries['Score'] = Countries['Score'] + Countries['Net Migration Rate/1000 Population'] + Countries['zLabor Force/Population']
Countries['Score'] = Countries['Score'] + Countries['Internet Users/Population']

#Compute Rank
Countries['Rank'] = Countries['Score'].rank(ascending=False)

#display result
result = Countries[['Country Name','Score','Rank']].sort(columns='Rank')
print(result)
```
